```python
import models1
from ensamble import Ensamble
from ensamble import Ensamble_diferente_peso_gradiente
import pandas as pd
import pickle
from tqdm import tqdm_notebook as tqdm
from sklearn.model_selection import train_test_split
```

```python
df_C_Soil = pd.read_csv('data/training_13C_suelos_v2.csv')
df_C_Plant = pd.read_csv('data/training_13C_plantas_v2.csv')

df_C_Soil = df_C_Soil.rename(columns={'13C': 'iso'})
df_C_Plant = df_C_Plant.rename(columns={'13C': 'iso'})
```

```python
data = []
n=1

for df_cuenca in [df_C_Soil, df_C_Plant]:
    y = df_cuenca['iso'].values
    dff = df_cuenca
    del dff['iso']
    X = dff[dff.keys()].values
    modelos = []
    for i in tqdm(range(0, 10)):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
        #model 1
        model1 = models1.ModelForest(X_train, y_train)
        model1.fit(10)
        modelos.append(model1)

        #model2
        model2 = models1.ModelMLP(X_train, y_train)
        model2.fit('tanh')
        modelos.append(model2)
    if n==1:
        c_soil = Ensamble(modelos)
        n=2
    else:
        c_plant = Ensamble(modelos)
    
    
```

```python
file_pi = open('c_soil.obj', 'wb')

pickle.dump(c_soil, file_pi)
```
